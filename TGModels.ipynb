```python
from matplotlib import pyplot as plt

ts = [0, 13, 20, 32, 42, 55, 65, 75, 85, 88, 95, 98, 107, 115, 120]
Vs = [250, 255, 550, 575, 576, 800, 1050, 1250, 1750, 2000, 2550, 2750, 3000, 3500, 4000]
```

Mendolsohn groei

c*V^d

```python
V = 250

#factors
c = 5
d = 0.8

start = 1
end = 120
step = 1

vdm = [V]
tdm = [start]

for timeperiod in range(start, end, step):
    tdm.append(timeperiod)
    V = c*pow(V,d)
    vdm.append(V)

plt.plot(tdm, vdm, "or")
plt.plot(ts, Vs, "ob")

```

c*v/v+d


```python
V = 250

#factors
c = 2
d = 1

start = 1
end = 120
step = 1

vdm = [V]
tdm = [start]

for timeperiod in range(start, end, step):
    tdm.append(timeperiod)
    V = c*(V/V+d)
    vdm.append(V)

plt.plot(tdm, vdm, "or")
plt.plot(ts, Vs, "ob")

```
