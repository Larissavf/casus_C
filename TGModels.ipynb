```python
from matplotlib import pyplot as plt

ts = [0, 13, 20, 32, 42, 55, 65, 75, 85, 88, 95, 98, 107, 115, 120]
Vs = [250, 255, 550, 575, 576, 800, 1050, 1250, 1750, 2000, 2550, 2750, 3000, 3500, 4000]
```

Mendolsohn groei

c*V^d

```python
V = 250

#factors
c = 5
d = 0.8

start = 1
end = 120
step = 1

vdm = [V]
tdm = [start]

for timeperiod in range(start, end, step):
    tdm.append(timeperiod)
    V = c*pow(V,d)
    vdm.append(V)

plt.plot(tdm, vdm, "or")
plt.plot(ts, Vs, "ob")

```

Lineair gelimiteerde groei

c*v/v+d


```python
V = 250

#factors
c = 2
d = 1

start = 1
end = 120
step = 1

vdm = [V]
tdm = [start]

for timeperiod in range(start, end, step):
    tdm.append(timeperiod)
    V = c*(V/V+d)
    vdm.append(V)

plt.plot(tdm, vdm, "or")
plt.plot(ts, Vs, "ob")

```

```python
import TGModels as models

model = models.Lineair_growth(250, 3)

vdm = [model.solve_rungekutta(i) for i in range(1,121)]

plt.plot(range(120), vdm, "or")
plt.plot(ts, Vs, "ob")
```

```python
model = models.Mendelsohn_growth(1)

model.fit(ts, Vs)

vdm = [model.solve_rungekutta(i) for i in range(120)]
plt.plot(range(120), vdm, "or")
plt.plot(ts, Vs, "ob")
```

```python
# 1. Importeer het model en maak een instantie
model = models.Mendelsohn_growth(c=1.5, d=0.8)

# 2. Controleer de parameters
print("Startparameters:", model.params())

# 3. Bereken de groeisnelheid voor een specifieke waarde van y
y_value = 10
growth_rate = model.formule(y_value)
print(f"Groeisnelheid bij y={y_value}: {growth_rate}")

# 4. Pas de parameters aan
model.setparams(c=2.0, d=0.9)
print("Aangepaste parameters:", model.params())

# 5. Fit het model met data
# Definieer tijdstappen (x) en waargenomen waarden (y)
time_steps = [1, 2, 3, 4, 5]
observed_values = [2.0, 3.8, 5.6, 7.2, 8.8]

model.fit(time_steps, observed_values)

# 6. Bekijk de geoptimaliseerde parameters
print("Geoptimaliseerde parameters:", model.params())
```
