## TGModels
Hier worden verschillende groei modellen gefit en getest op TumorGrowth data, welke het beste past.
De volgende modellen:
    - lineair growth  
    - allee effect growth  
    - exponential decreasing growth  
    - exponential increasing growth  
    - grompertz growth  
    - mendelsohn growth  
    - surface limited growth  
    - von bertalanffy growth  



Het wordt opgelost doormiddel van de Runge Kutta methode of de Euler's methode.  
Vervolgens getest met de BIC waarde.

#### Data

```python
# data
data_ts = [
    3.46,  4.58,  5.67,  6.64,  7.63,  8.41,  9.32, 10.27, 11.19,
    12.39, 13.42, 15.19, 16.24, 17.23, 18.18, 19.29, 21.23, 21.99,
    24.33, 25.58, 26.43, 27.44, 28.43, 30.49, 31.34, 32.34, 33.00,
    35.20, 36.34, 37.29, 38.50, 39.67, 41.37, 42.58, 45.39, 46.38,
    48.29, 49.24, 50.19, 51.14, 52.10, 54.00, 56.33, 57.33, 59.38,
]
data_ys = [
    0.0158, 0.0264, 0.0326, 0.0445, 0.0646, 0.0933, 0.1454, 0.2183, 0.2842,
    0.4977, 0.6033, 0.8441, 1.2163, 1.4470, 2.3298, 2.5342, 3.0064, 3.4044,
    3.2046, 4.5241, 4.3459, 5.1374, 5.5376, 4.8946, 5.0660, 6.1494, 6.8548,
    5.9668, 6.6945, 6.6395, 6.8971, 7.2966, 7.2268, 6.8815, 8.0993, 7.2112,
    7.0694, 7.4971, 6.9974, 6.7219, 7.0523, 7.1095, 7.0694, 8.0562, 7.2268, 
]
```

```python
# schalen
import numpy as np
# Scale the data_ts to [-1, 1]
data_ts = np.array(data_ts)
min_val = np.min(data_ts)
max_val = np.max(data_ts)
scaled_ts = 2 * (data_ts - min_val) / (max_val - min_val) - 1
```

#### Models

```python
# models
import TGModels as model
x = scaled_ts
y = data_ys


models = [
    model.Lineair_growth,
    model.Exponentieel_toenemende_groei,
    model.Exponentieel_afvlakkende_groei,
    model.Mendelsohn_growth,
    model.Allee_effect_growth,
    model.Lineair_gelimiteerde_groei,
    model.surface_limited_growth,
    model.Von_Bertalanffy_groei,
    model.Gompertz_growth
]
```

#### Models fitten

Alle bovenstaande beschikbare models zijn toegepast in de onderstaande cellen. Hierin zijn de modellen gefit, en geplot op de data.

```python
import matplotlib.pyplot as plt
```

```python
model_lin = model.Lineair_growth(1)
model_lin.fit(x, y)

ts = np.linspace(-1, 1, 600)  # Scaled time series for model
y_model = [model_lin.solve_rungekutta(i) for i in ts]
plt.plot(ts, y_model, "-r")
plt.plot(x, y, "ob")
plt.show
```

```python
model_ext = model.Exponentieel_toenemende_groei(1)
model_ext.fit(x, y)

ts = np.linspace(-1, 1, 600)  # Scaled time series for model
y_model = [model_ext.solve_rungekutta(i) for i in ts]
plt.plot(ts, y_model, "-r")
plt.plot(x, y, "ob")
plt.show
```

### Models testen

Je kan de models testen met gebruik van de BIC waard. Des de lager het getal des te beter de fit is. 

In dit voorbeeld is dus Gompertz Growth de beste model is voor deze data.

```python
# bic waarde
import pandas as pd

pd.DataFrame({"BIC" : [model_lin.BIC(x,y), 
                       model_all.BIC(x,y),
                       model_exa.BIC(x,y),
                       model_ext.BIC(x,y),
                       model_Gom.BIC(x,y),
                       model_men.BIC(x,y),
                       model_sur.BIC(x,y),
                       model_vbe.BIC(x,y)]}, index = ["Lineair growth", "Allee effect growth", 
                                                      "Exponentieel decreasing growth", "Exponentieel increasing growth", 
                                                      "Gompertz growth", "Mendelsohn growth", "surface limited growth", "Von Bertalanffy growth"]).sort_values('BIC')
```
